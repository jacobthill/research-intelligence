```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib_venn import venn2, venn2_circles
```

## Testing different search strategies

Two csv files are provided as output: 'som_gi_report.csv' and 'som_gi_report_no_dupes.csv'. The first is a report of all publications. In cases where researchers from SOM GI co-authored a publication, the publication will be recorded once for each author. The second has duplications removed so each publication is recorded only once. It can be used to track total publications for the department. A third file 'som_gi_report_all.csv' can be used to evaluate the various search strategies used. It includes many duplicate publications. If a paper was co-authored by 2 SOM GI researchers it could appear 0 to 8 (2 researchers x 4 search strategies) times in the results.

```python
som_dataset = pd.read_csv('output/som_gi_report_all.csv',delimiter=',')
```

### 1. Searching by name string and GRID

The initial search strategy uses a name string (e.g. 'Afrin Kamal') and a list of GRID ids fetched from an approximate match of the institution name string; this is the recommended approach to search for institutions in Dimensions.

### 2. Searching by Dimensions researcher ID and GRID

In the same way the institution name string can be used to fectch multiple GIRD ids, the researcher name string can be used to fetch 0 or multiple researcher ids; these are ids assignment by Dimensions and used to disambiguate researchers.

### 3. Searching by ORCID

It is also possible to search for publications with a matching ORCID.

### 4. Searching by Dimensions researcher ID only

Finally it is possible to use the name string to fetch a list of Dimensions researcher ids (as done above) without incorporating the institution field. This approach has the advantage of returing results when alternate instituitons have not been fully added to the input data or in publications where Dimensions has left this field blank. In cases where the researcher name is uncommon, it may be a useful addition to the strategies above. In cases where the researcher name is common, this approach has the potential to overwhelm the results with a significant percentage of false positives. 

### Searching by name string and GRID (1): Results

```python
name_string_and_grid = som_dataset[(som_dataset['search_strategy'] == 'Name string & GRID')]
not_name_string_and_grid = som_dataset[(som_dataset['search_strategy'] == 'ORCID') | (som_dataset['search_strategy'] == 'Reasearcher id & GRID')]

# plt.figure(figsize=(4,4))
set1 = set(name_string_and_grid['dimensions_id'])
set2 = set(not_name_string_and_grid['dimensions_id'])

print("Name string & GRID: {}".format(len(set1)))
print("Other (Researcher id & GRID, or ORCID): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('Name string & GRID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

For compasison we can also plot the Venn diagram with results from searching only the researcher id included.

```python
name_string_and_grid = som_dataset[(som_dataset['search_strategy'] == 'Name string & GRID')]
not_name_string_and_grid = som_dataset[(som_dataset['search_strategy'] != 'Name string & GRID')]

# plt.figure(figsize=(4,4))
set1 = set(name_string_and_grid['dimensions_id'])
set2 = set(not_name_string_and_grid['dimensions_id'])

print("Name string & GRID: {}".format(len(set1)))
print("Other (Researcher id & GRID, ORCID, or researcher id alone): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('Name string & GRID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

```python
researcher_id_and_grid = som_dataset[(som_dataset['search_strategy'] == 'Reasearcher id & GRID')]
not_researcher_id_and_grid = som_dataset[(som_dataset['search_strategy'] == 'ORCID') | (som_dataset['search_strategy'] == 'Name string & GRID')]

# plt.figure(figsize=(4,4))
set1 = set(researcher_id_and_grid['dimensions_id'])
set2 = set(not_researcher_id_and_grid['dimensions_id'])

print("Researcher id & GRID: {}".format(len(set1)))
print("Other (Name string & GRID, or ORCID): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('Researcher Id & GRID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

Searching by researcher id seems to have a small effect on the results; it returned 52 publications that were not returned by the other approaches. Again we can compare this with the results of the research id alone (strategy 4) included. 

```python
researcher_id_and_grid = som_dataset[(som_dataset['search_strategy'] == 'Reasearcher id & GRID')]
not_researcher_id_and_grid = som_dataset[(som_dataset['search_strategy'] != 'Reasearcher id & GRID')]

# plt.figure(figsize=(4,4))
set1 = set(researcher_id_and_grid['dimensions_id'])
set2 = set(not_researcher_id_and_grid['dimensions_id'])

print("Researcher id & GRID: {}".format(len(set1)))
print("Other (Name string & GRID, ORCID, or researcher id alone): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('Researcher Id & GRID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

As expected the set of publications returned from searching the research id and GRID (strategy 2) is completely subsummed withing the set of searching the researcher id alone (strategy 4).

### Searching by ORCID (3): Results

One question that we will want to track over time is ORCID coverage. It is expected that ORCID will, in time, eliminate name ambiguity problems. The current state of ORCID coverage is unclear. To get a sense of ORCID coverage in Dimensions we can plot a Venn diagram of two sets of returned publications: those retrieved from an ORCID query, and those retreived from either a name string search and a GRID or a research id search and a GRID. Results returned from a research id alone are excluded because of the high likelihood of false positives. 

```python
orcid = som_dataset[(som_dataset['search_strategy'] == 'ORCID')]
not_orcid = som_dataset[(som_dataset['search_strategy'] == 'Name string & GRID') | (som_dataset['search_strategy'] == 'Reasearcher id & GRID')]

# plt.figure(figsize=(4,4))
set1 = set(orcid['dimensions_id'])
set2 = set(not_orcid['dimensions_id'])

print("Orcid: {}".format(len(set1)))
print("Other (Name string & GRID, or ORCID): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('ORCID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

For compasison we can also plot the Venn diagram with results from searching only the researcher id included.

```python
orcid = som_dataset[(som_dataset['search_strategy'] == 'ORCID')]
not_orcid = som_dataset[(som_dataset['search_strategy'] != 'ORCID')]

# plt.figure(figsize=(4,4))
set1 = set(orcid['dimensions_id'])
set2 = set(not_orcid['dimensions_id'])

print("Orcid: {}".format(len(set1)))
print("Other (Name string & GRID, ORCID, or researcher id alone): {}".format(len(set2)))
print("Total: {}".format(len(set1.difference(set2)) + len(set2.difference(set1)) + len(set1.intersection(set2))))

plt.figure(figsize=(6,4))
v = venn2(subsets = (len(set1.difference(set2)),
                     len(set2.difference(set1)),
                     len(set1.intersection(set2))), set_labels = ('ORCID', 'Other'))
           
c = venn2_circles(subsets = (len(set1.difference(set2)),
                             len(set2.difference(set1)),
                             len(set1.intersection(set2))), linestyle='solid')

plt.show()
```

ORCID is clearly not sufficiently covered in Dimensions to use it as our sole search strategy, however, it is clearly useful as an additive approach. It was able to return unique results beyond what the other three strategies could return. ~4% of the publications could only be returned via an ORCID query. 

By comparing the two Venn diagrams we can also see that searching by researcher ID alone added 2,401 results that were not returned by any other approach. The assumption we are opterating on is that these results will include many false positives because of the likelihood of many researchers (in the whole world) having the same name. It is impossible to know who noisy the data will be without a manual inspection but we can assume that at least 729 of these 2,401 results (956-227: the intersection between the ORCID and research id only results) are correct.

Future inquiries could attempt to understand the current limitations of ORCID by looking at coverage over time (Are recent publicaitons better covered than others?), by researcher (Are some researchers well covered while others are not covered at all?), or by field (Are some research areas better represented than others? If so, which ones?) A better understanding of the current coverage limitations could help us determine campus wide strategies to improve ORCID coverage.

### Summary

It will be useful to take some time to consider the results in the context of other reports before drawing major conclusions but, at the moment, it seems safe to say that multiple search strategies are necessary. 

It is also clear that strategies 2 and 4 are mutually exclusive; the results of 2 will always be contained in 4. Which strategy is preferable depends on the willingness of the stakeholder to do manual cleaning of the results. Do we want a highly precise report with some missing publications or is a potentially high number of false positives acceptable? Of course, uncommon names will have fewer false positives. Some researchers may not even notice the additional noise and may welcome strategy 4. Others may find that the majority of the results are false positives. In reality strategy 4 is probably not a viable option for all cases; it probably couldn't be implemented in Profiles, for example. But some departments may want to use this strategy for internal reports. 

We can also say that ORCID is highly useful as an additive search strategy but cannot stand alone as the sole strategy we use. More analysis needs to be done to determine whether the gaps in coverage correlates with date, research area, or researcher behavior (e.g. adoption and use of ORCID IDs). ORCID search should be added to the SUL-Pub harvesting code whether we continue to harvest from Web of Science subscription or shift to Dimensions.
